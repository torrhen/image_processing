### Shading correction using image multiplication

Given a perfect image $f(x, y)$ and a shading function $h(x, y)$ to produce an output image given by $g(x, y) = f(x, y)h(x, y)$, the output image and an estimation of the shading function can be used to find the original input $f(x, y)$.

The input image can be found by multiplying the output image by the inverse of the shading function.

$f(x, y) = g(x, y) \times \frac{1}{h(x, y)}$

Even if the shading function is not known, we can use an estimation to find good reconstructions for the input.

```python
# read in shaded input image 
image = cv.imread("shaded_image.png")
# drop last column so images are the same shape
image = image[:, :-1, :]
# display shaded image
cv2_imshow(image)
```

```python
# read in approximation of shading function
shading_function = cv.imread("shading_function.png")
# drop last row so images are the same shape
shading_function = shading_function[:-1, :, :]
# display shading function
cv2_imshow(shading_function)

print(shading_function.shape)
```

```python
# recreate original input by multiplying shaded input by inverse of shaded function
input = (image / shading_function) * 255
# display original input image
cv2_imshow(input)
```
