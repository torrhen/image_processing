```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

### Region of Interest (RoI) masking using image multiplication

```python
# read in example greyscale input image
input = cv.imread("highway.jpg")
input = cv.cvtColor(input, cv.COLOR_BGR2GRAY)
# display input image
cv2_imshow(input)
```

```python
# create an RoI mask for isolating pixel regions in original input
roi_mask = np.zeros((input.shape[0], input.shape[1]))
# position 2 50 x 50 regions of 1's 
for i in range(roi_mask.shape[0]):
  for j in range(roi_mask.shape[1]):
    if (i >= 50 and i < 100) and (j >= 50 and j < 100):
      roi_mask[i][j] = 1
    if (i >= 100 and i < 150) and (j >= 200 and j < 250):
      roi_mask[i][j] = 1 
# display RoI mask
cv2_imshow(roi_mask * 255)
```

```python
# create new image by masking the input image
output = input * roi_mask
# display new image
cv2_imshow(output)
```
