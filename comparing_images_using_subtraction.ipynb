Image subtraction is used to enhance the differences between images.

```python
import cv2 as cv
import numpy as np
import requests
from google.colab.patches import cv2_imshow

# display greyscale image
image = cv.imread("highway.jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(image)
```

```python
# make changes to a opy of the original image
image_copy = image.copy()
# set pixels >= 180 to white
image_copy[image_copy >= 180] = 255
# display modified image
cv2_imshow(image_copy)
```

```python
# enhance pixels >= 180 in original image using image subtraction
image_enhanced = image - image_copy
# display enhanced image
cv2_imshow(image_enhanced)
```

### Colour

```python
# display original BGR image
image_colour = cv.imread("highway.jpg")
cv2_imshow(image_colour)
```

```python
# convert original BGR image to HSV colourspace
image_hsv = cv.cvtColor(image_colour, cv.COLOR_BGR2HSV)
cv2_imshow(image_hsv)
```

```python
# make changes to a opy of the original image
image_hsv_copy = image_hsv.copy()
# set value pixels >= 180 to white
image_hsv_copy = image_hsv_copy.reshape(-1, 3)
for i in range(image_hsv_copy.shape[0]):
  if image_hsv_copy[i][-1] >= 180:
    image_hsv_copy[i][-1] = 255
# reshape into image
image_hsv_copy = image_hsv_copy.reshape(image_colour.shape[0], image_colour.shape[1], -1)

# display modified image
cv2_imshow(image_hsv_copy)
```

```python
# enhance value pixels >= 180 in original image using image subtraction
image_hsv_enhanced = image_hsv - image_hsv_copy
# display enhanced image
cv2_imshow(image_hsv_enhanced)
```

```python
# display enhanced image in BGR colourspace
image_BGR_enhanced = cv.cvtColor(image_hsv_enhanced, cv.COLOR_HSV2BGR)
cv2_imshow(image_BGR_enhanced)
```
